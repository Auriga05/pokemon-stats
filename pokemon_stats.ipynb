# subplots with histograms

```python
from plotly import tools
temp_data = data
total = temp_data['TOTAL']
totz = total.tolist()
trace0=go.Histogram(x=totz)

hp=temp_data['HP']
hpz = hp.tolist()
trace1=go.Histogram(x=hpz)

attack=temp_data['ATTACK']
attackz = attack.tolist()
trace2=go.Histogram(x=attackz)

speed=temp_data['SPEED']
speedz = speed.tolist()
trace3=go.Histogram(x=speedz)

defense=temp_data['DEFENSE']
defensez = defense.tolist()
trace4=go.Histogram(x=defensez)
```

```python
fig = tools.make_subplots(rows=3, cols=2, specs=[[{'colspan': 2}, None],[{}, {}],[{}, {}]],
                          subplot_titles=('Histogram Of Total','Histogram Of HP', 'Histogram Of Attack','Histogram Of Speed', 'Histogram Of Defense'))

fig.append_trace(trace0, 1, 1)
fig.append_trace(trace1, 2, 1)
fig.append_trace(trace2, 2, 2)
fig.append_trace(trace3, 3, 1)
fig.append_trace(trace4, 3, 2)


fig['layout'].update(showlegend=False, title='Histogram Subplots')
py.iplot(fig, filename='custom-sized-subplot-with-subplot-titles')
```

# box plots of all the stats 

```python
HP = data['HP']
SPEED = data['SPEED']
ATTACK = data['ATTACK']
DEFENSE = data['DEFENSE']
SPATTACK = data['SP.ATK']
SPDEFENSE = data['SP.DEF']
TOTAL = data['TOTAL']
```

```python
hp = HP.tolist()
speed = SPEED.tolist()
attack = ATTACK.tolist()
defense = DEFENSE.tolist()
spattack = SPATTACK.tolist()
spdefense = SPDEFENSE.tolist()
totall = TOTAL.tolist()
```

```python
trace0 = go.Box(
    y=hp,
    name = 'HP',
)

trace1 = go.Box(
    y=speed,
    name = 'SPEED',
)

trace2 = go.Box(
    y=attack,
    name = 'ATTACK',
)

trace3 = go.Box(
    y=defense,
    name = 'DEFENSE',
)

trace4 = go.Box(
    y=spattack,
    name = 'SPECIAL ATTACK',
)

trace5 = go.Box(
    y=spdefense,
    name = 'SPECIAL DEFENSE',
)

trace6 = go.Box(
    y=totall,
    name = 'TOTAL',
)

py.iplot([trace0, trace1, trace2, trace3, trace4, trace5, trace6],filename = "boxPlot")
```
